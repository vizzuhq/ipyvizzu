```python
###( load dataset 1 )###
eu_df = pd.read_csv(
    "https://ipyvizzu.vizzuhq.com/0.17/assets/data/chart_types_eu.csv",
    dtype={
        "Year": str, 
        "Timeseries": str
    },
)
data_eu = Data()
data_eu.add_df(eu_df)


###( load dataset 2 )###
eu_6_df = pd.read_csv(
    "https://ipyvizzu.vizzuhq.com/0.17/assets/data/chart_types_eu_data_6.csv",
    dtype={
        "Year": str, 
        "Timeseries": str
    },
)
data_eu_6 = Data()
data_eu_6.add_df(eu_6_df)


###( load dataset 3 )###
eu_3_df = pd.read_csv(
    "https://ipyvizzu.vizzuhq.com/0.17/assets/data/chart_types_eu_data_3.csv",
    dtype={
        "Year": str, 
        "Timeseries": str
    },
)
data_eu_3 = Data()
data_eu_3.add_df(eu_3_df)


###( load dataset 4 )###
eu_4_df = pd.read_csv(
    "https://ipyvizzu.vizzuhq.com/0.17/assets/data/chart_types_eu_data_4.csv",
    dtype={
        "Year": str, 
        "Timeseries": str
    },
)
data_eu_4 = Data()
data_eu_4.add_df(eu_4_df)


###( load dataset 8 )###
eu_8_df = pd.read_csv(
    "https://ipyvizzu.vizzuhq.com/0.17/assets/data/chart_types_eu_data_8.csv",
    dtype={
        "Year": str, 
        "Timeseries": str
    },
)
data_eu_8 = Data()
data_eu_8.add_df(eu_8_df)



###( load dataset Music Industry History - 1 )###
mih_1_df = pd.read_csv(
    "https://ipyvizzu.vizzuhq.com/0.17/assets/data/music_industry_history_1.csv",
    dtype={
        "Year": str, 
        "Timeseries": str
    },
)
data_mih_1 = Data()
data_mih_1.add_df(mih_1_df)
```

```python
###( Drill down: Area Chart )###
### https://ipyvizzu.vizzuhq.com/latest/examples/analytical_operations/drilldown/area/

chart = Chart()
chart.animate(data_eu_6)
chart.animate(
    Config(
        {
            "channels": {"x": "Year", "y": "Value 2 (+)"},
            "geometry": "area",
            "title": "Drill down: Area Chart",
        }
    )
)
 
chart.animate(
    Config(
        {
            "channels": {
                "y": ["Value 2 (+)", "Country"],
                "color": "Country",
            }
        }
    )
)
 
chart.feature("tooltip", True)
```

```python
###( Drill down: Polar Area )###
### https://ipyvizzu.vizzuhq.com/latest/examples/analytical_operations/drilldown/area_polar/

chart = Chart()
chart.animate(data_eu_6)
chart.animate(
    Config(
        {
            "channels": {"x": "Year", "y": "Value 2 (+)"},
            "geometry": "area",
            "coordSystem": "polar",
            "title": "Drill down: Polar Area",
        }
    )
)
 
chart.animate(
    Config(
        {
            "channels": {
                "y": ["Value 2 (+)", "Country"],
                "color": "Country",
            }
        }
    )
)
 
chart.feature("tooltip", True)
```

```python
###( Drill Down: Bubble and Distribution )###
### https://ipyvizzu.vizzuhq.com/latest/examples/analytical_operations/drilldown/bubble_and_distribution/

chart = Chart()
chart.animate(data_eu_8)
chart.animate(
    Config(
        {
            "channels": {
                "color": "Country",
                "size": "Value 2 (+)",
                "label": "Value 2 (+)",
            },
            "geometry": "circle",
            "title": "Drill down: Bubble and Distribution",
        }
    )
)
 
chart.animate(
    Config(
        {
            "channels": {"size": ["Year", "Value 2 (+)"]},
            "geometry": "circle",
        }
    )
)
 
chart.animate(
    Config(
        {
            "channels": {
                "x": "Value 1 (+)",
                "y": "Value 3 (+)",
                "noop": "Year",
                "size": "Value 2 (+)",
                "label": None,
            }
        }
    )
)
 
chart.feature("tooltip", True)
```

```python
###( Drill Down: Column 1 )###
### https://ipyvizzu.vizzuhq.com/latest/examples/analytical_operations/drilldown/column_1/

chart = Chart()
chart.animate(data_eu_4)
chart.animate(
    Config(
        {
            "channels": {
                "y": "Value 2 (+)", 
                "label": "Value 2 (+)"
            },
            "title": "Drill down: Column 1",
        }
    )
)
 
chart.animate(
    Config(
        {
            "channels": {
                "x": ["Year", "Country"],
                "y": "Value 2 (+)",
                "color": "Country",
                "label": None,
            }
        }
    )
)
 
chart.feature("tooltip", True)
```

```python
###( Drill Down: Column 2 )###
### https://ipyvizzu.vizzuhq.com/latest/examples/analytical_operations/drilldown/column_2/

chart = Chart()
chart.animate(data_eu_4)
chart.animate(
    Config({"channels": {"y": "Value 2 (+)", "label": "Value 2 (+)"},   
            "title": "Drill down: Column 2",})
)
 
chart.animate(
    Config(
        {
            "channels": {
                "x": "Year",
                "y": ["Country", "Value 2 (+)"],
                "color": "Country",
                "label": None,
            }
        }
    )
)
 
chart.feature("tooltip", True)
```

```python
###( Drill Down: Column 3 )###
### https://ipyvizzu.vizzuhq.com/latest/examples/analytical_operations/drilldown/column_3/

chart = Chart()
chart.animate(data_eu_8)
chart.animate(
    Config({"channels": {"y": "Value 2 (+)", "label": "Value 2 (+)"},
            "title": "Drill down: Column 3",})
)
 
chart.animate(Config({"channels": {"x": "Country"}}))
 
chart.feature("tooltip", True)
```

```python
###( Drill Down: Column 4 )###
### https://ipyvizzu.vizzuhq.com/latest/examples/analytical_operations/drilldown/column_4/

chart = Chart()
chart.animate(data_eu_8)
chart.animate(
    Config(
        {
            "channels": {
                "x": "Country",
                "y": "Value 2 (+)",
                "label": "Value 2 (+)",
            },
             "title": "Drill down: Column 4",
        }
    )
)
 
chart.animate(
    Config(
        {
            "channels": {
                "y": ["Value 2 (+)", "Joy factors"],
                "color": "Joy factors",
            }
        }
    ),
    Style(
        {
            "plot": {
                "marker": {
                    "colorPalette": "#ef675aFF #6d8cccFF #e6cf99FF #9c50abFF"
                }
            }
        }
    ),
)
 
chart.feature("tooltip", True)
```

```python
###( Drill Down: Stacked Column )###
### https://ipyvizzu.vizzuhq.com/latest/examples/analytical_operations/drilldown/column_stacked/

chart = Chart()
chart.animate(data_eu_8)
chart.animate(
    Config(
        {
            "channels": {
                "y": ["Country", "Value 2 (+)"],
                "color": "Country",
                "label": "Value 2 (+)",
            },
            "title": "Drill down: Stacked Column",
        }
    )
)
 
chart.animate(Config({"channels": {"x": "Year", "label": None}}))
 
chart.feature("tooltip", True)
```

```python
###( Drill Down: Donut )###
### https://ipyvizzu.vizzuhq.com/latest/examples/analytical_operations/drilldown/donut/

chart = Chart()
chart.animate(data_eu_8)
chart.animate(
    Config(
        {
            "channels": {
                "x": ["Country", "Value 2 (+)"],
                "y": {"range": {"min": "-200%"}},
                "color": "Country",
                "label": "Value 2 (+)",
            },
            "coordSystem": "polar",
            "title": "Drill down: Donut",
        }
    )
)
 
chart.animate(
    Config(
        {
            "channels": {
                "x": ["Year", "Value 2 (+)"],
                "y": {"set": "Country", "range": {"min": "-30%"}},
                "label": None,
            }
        }
    )
)
 
chart.animate(
    Config(
        {
            "channels": {
                "x": "Year",
                "y": {
                    "set": ["Country", "Value 2 (+)"],
                    "range": {"min": "-30%"},
                },
            }
        }
    ),
    Style({"plot": {"marker": {"rectangleSpacing": "0.1em"}}}),
)
 
chart.feature("tooltip", True)
```
