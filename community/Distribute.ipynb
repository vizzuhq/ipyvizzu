```python
###( Distribute: Stacked Area )###
### https://ipyvizzu.vizzuhq.com/latest/examples/analytical_operations/distribute/existingmeasure_area_stacked/

chart = Chart()
chart.animate(data_eu_6)
chart.animate(
    Config(
        {
            "channels": {
                "x": "Year",
                "y": ["Country", "Value 2 (+)"],
                "color": "Country",
            },
            "geometry": "area",
            "title": "Distribute: Stacked Area",
        }
    )
)
 
chart.animate(
    Config({"channels": {"y": "Value 2 (+)"}, "geometry": "circle"})
)
 
chart.feature("tooltip", True)
```

```python
###( Distribute: Bubble )###
### https://ipyvizzu.vizzuhq.com/latest/examples/analytical_operations/distribute/existingmeasure_bubble/

chart = Chart()
chart.animate(data_eu_14)
chart.animate(
    Config(
        {
            "channels": {
                "noop": ["Year", "Country"],
                "size": "Value 2 (+)",
            },
            "geometry": "circle",
            "title": "Distribute: Bubble",
        }
    )
)
 
chart.animate(
    Config(
        {
            "channels": {
                "noop": ["Year", "Country"],
                "y": "Value 2 (+)",
                "size": None,
            }
        }
    )
)
```

```python
###( Distribute: Stacked Bubble 1 )###
### https://ipyvizzu.vizzuhq.com/latest/examples/analytical_operations/distribute/existingmeasure_bubble_stacked_1/

chart = Chart()
chart.animate(data_eu_4)
chart.animate(
    Config(
        {
            "channels": {
                "color": "Country",
                "size": ["Year", "Value 2 (+)"],
                "label": "Value 2 (+)",
            },
            "geometry": "circle",
        }
    )
)
 
chart.animate(
    Config(
        {
            "channels": {
                "x": "Year",
                "y": {
                    "set": ["Country", "Value 2 (+)"],
                    "range": {"min": "-20%"},
                },
                "size": None,
                "label": None,
            },
            "geometry": "rectangle",
            "coordSystem": "polar",
        }
    ),
    Style({"plot": {"marker": {"rectangleSpacing": "0.1em"}}}),
)
 
chart.feature("tooltip", True)
```

```python
###( Distribute: Scatter Plot )###
### https://ipyvizzu.vizzuhq.com/latest/examples/analytical_operations/distribute/existingmeasure_scatterplot/

chart = Chart()
chart.animate(data_eu)
chart.animate(
    Config(
        {
            "channels": {
                "x": "Value 5 (+/-)",
                "y": "Value 6 (+/-)",
                "noop": "Joy factors",
                "lightness": "Year",
            },
            "geometry": "circle",
            "title": "Distribute: Scatter Plot",
        }
    )
)
 
chart.animate(
    Config({"channels": {"x": "Year"}, "orientation": "horizontal"})
)
 
chart.animate(Config({"channels": {"x": "Value 5 (+/-)"}}))
 
chart.animate(
    Config(
        {
            "channels": {"y": "Joy factors", "noop": None},
            "orientation": "vertical",
        }
    )
)
 
chart.animate(
    Config(
        {"channels": {"y": "Value 6 (+/-)", "noop": "Joy factors"}}
    )
)
 
chart.feature("tooltip", True)
```

```python
###( Distribute: Split Scatter Plot )###
### https://ipyvizzu.vizzuhq.com/latest/examples/analytical_operations/distribute/existingmeasure_scatterplot_split/

chart = Chart()
chart.animate(data_eu_6)
chart.animate(
    Config(
        {
            "channels": {
                "x": ["Value 1 (+)", "Country"],
                "y": "Value 3 (+)",
                "noop": "Year",
                "color": "Country",
            },
            "geometry": "circle",
            "split": True,
            "orientation": "vertical",
            "title": "Distribute: Split Scatter Plot",
        }
    )
)
 
chart.animate(
    Config({"channels": {"x": "Value 1 (+)"}, "split": False})
)
 
chart.feature("tooltip", True)
```
