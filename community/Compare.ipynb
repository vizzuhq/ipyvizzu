```python
###( Compare: 100% Stacked Area )###
### https://ipyvizzu.vizzuhq.com/latest/examples/analytical_operations/compare/area_100percent_stacked/

chart = Chart()
chart.animate(data_eu_6)
chart.animate(
    Config(
        {
            "channels": {
                "x": "Year",
                "y": ["Value 2 (+)", "Country"],
                "color": "Country",
            },
            "geometry": "area",
            "align": "stretch",
            "title": "Compare: 100% Stacked Area",
        }
    )
)
 
chart.animate(
    Config(
        {
            "channels": {"y": "Value 2 (+)"},
            "geometry": "line",
            "align": "none",
        }
    )
)
 
chart.feature("tooltip", True)
```

```python
###( Compare: Stacked Stream )###
### https://ipyvizzu.vizzuhq.com/latest/examples/analytical_operations/compare/stream_stacked/

chart = Chart()
chart.animate(data_mih_1)
method = """
            const year = parseFloat(event.detail.text)
            if (!event.detail.text.includes('$') && !isNaN(year) && year % 5 !== 0)
                event.preventDefault()
        """
handler = chart.on("plot-axis-label-draw", method)
 
chart.animate(
    Config(
        {
            "channels": {
                "x": "Year",
                "y": ["Revenue", "Format"],
                "color": "Format",
            },
            "geometry": "area",
            "align": "center",
            "title": "Compare: Stacked Stream",
        }
    ),
    Style(
        {"plot": {"yAxis": {"label": {"numberScale": "K, M, B, T"}}}}
    ),
)
 
chart.animate(Config({"align": "none", "split": True}))
 
chart.animate(
    Config(
        {
            "channels": {"x": ["Format", "Year"], "y": "Revenue"},
            "split": False,
        }
    )
)
 
chart.feature("tooltip", True)
```
